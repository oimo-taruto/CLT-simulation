```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

Apple=yf.Ticker("AAPL")

Apple_earnings_dates=Apple.get_earnings_dates()
Apple_earnings_dates_list=[date.date() for date in Apple_earnings_dates.index]

all_data=yf.download(tickers='AAPL',period='max')
earnings_dates = pd.to_datetime(Apple_earnings_dates_list)
earnings_data=all_data.loc[all_data.index.isin(Apple_earnings_dates_list)]

df=earnings_data
df=df.reset_index().copy()
df.columns=['Date','Close','Open','High','Low','Volume']
df
```
