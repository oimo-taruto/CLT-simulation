```python
#準備と母集団の定義
import numpy as np
#数値計算のライブラリの読み込み
import matplotlib.pyplot as plt
#データの可視化　描画モジュールの読み込み

population=np.arange(1,7)
#母集団の定義　今回はサイコロの目の離散一様分布
print(f"母集団の要素:{population}")

sample_sizes=[10,100,500]
#標本サイズのリスト　これが大きいと正規分布に近似→CLT
num_samples=5000
#標本抽出と平均計算を繰り返す回数
#大きいと図が滑らかになるが、処理に時間がかかる

plt.figure(figsize=(15,5))

#シミュレーションの実行
for i,n in enumerate(sample_sizes):
#標本サイズリストの各要素に対して繰り返し処理を行う
#enumerate関数は、リストの各要素に0から番号(インデックス i)を振ってくれる
#iがインデックス、nが要素の値(今回は、2,10,30)を表す
  sample_means=[]
  #各nに対する標本平均を格納するための空のリスト

  for _ in range(num_samples):
  #5000回内側の処理(標本抽出と平均計算)を繰り返す
    sample=np.random.choice(population,size=n)
    #母集団から復元抽出でn個の標本sampleを抽出
    sample_mean=np.mean(sample)
    #抽出したn個の標本の平均を計算
    sample_means.append(sample_mean)
    #計算した標本平均をリストに格納

#結果の確認と可視化
  plt.subplot(1,3,i+1)
  #グラフを1行3列に配置し、現在の繰り返し回数に応じて描画位置を指定
  plt.hist(sample_means,bins=20,edgecolor='black',color='skyblue',density=True)
  #ヒストグラムを描写
  #destiny=True→ヒストグラムの面積の合計を1にする(確率密度表示)
  plt.title(f'SAmple Size(n)={n}')
  plt.xlabel('Sample Mean')
  plt.ylabel('Density')

plt.suptitle(f'Distribution of Sample Means(Central Limit Theorem) from Uniform Distribution,Repeats:{num_samples}',fontsize=14)
plt.tight_layout(rect=[0,0.03,1,0.95])
plt.show()
#グラフを表示
```
