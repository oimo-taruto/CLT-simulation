```python
import numpy as np
import matplotlib.pyplot as plt
population=np.arange(1,7)
print(f"母集団の要素:{population}")

sample_sizes=[2,10,30]
num_samples=5000

plt.figure(figsize=(15,5))

for i,n in enumerate(sample_sizes):
  sample_means=[]

  for _ in range(num_samples):
    sample=np.random.choice(population,size=n)
    sample_mean=np.mean(sample)
    sample_means.append(sample_mean)

  plt.subplot(1,3,i+1)
  plt.hist(sample_means,bins=20,edgecolor='black',color='skyblue',density=True)
  plt.title(f'SAmple Size(n)={n}')
  plt.xlabel('Sample Mean')
  plt.ylabel('Density')

plt.suptitle(f'Distribution of Sample Means(Central Limit Theorem) from Uniform Distribution,Repeats:{num_samples}',fontsize=14)
plt.tight_layout(rect=[0,0.03,1,0.95])
plt.show()
```
