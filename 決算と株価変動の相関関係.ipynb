## Extract Earnings Dates

### Subtask:
Extract individual earnings dates from the `Apple_earnings_dates` DataFrame.


**Reasoning**:
The previous code failed due to an incorrect `axis` parameter used with `min()` and `max()` on a 1-dimensional index. I need to correct this error and extract the date components from the `Apple_earnings_dates` DataFrame's index and store them as a list of `datetime.date` objects.



```python
import yfinance as yf
import pandas as pd

Apple=yf.Ticker("AAPL")
Microsoft=yf.Ticker("MSFT")
Alphabet=yf.Ticker("GOOGL")
Amazon=yf.Ticker("AMZN")
Meta=yf.Ticker("META")

Apple_earnings_dates=Apple.get_earnings_dates()
df_Apple=pd.DataFrame(Apple_earnings_dates)

# Extract start and end dates from the DataFrame index
# Corrected: .min() and .max() are applied directly on the index without an axis parameter.
start_date = Apple_earnings_dates.index.min().date()
end_date = Apple_earnings_dates.index.max().date()

Apple_fl=Apple.history(start=start_date, end=end_date)

# Extract individual earnings dates into a list of datetime.date objects
earnings_dates_list = [date.date() for date in Apple_earnings_dates.index]

print("df_Apple DataFrame:\n", df_Apple)
print("\nApple_fl DataFrame:\n", Apple_fl)
print("\nExtracted Earnings Dates (list):\n", earnings_dates_list)
```

## Fetch History for Each Earnings Date

### Subtask:
For each earnings date, retrieve 2 days of stock history (the earnings date itself and the following day) using the `Apple.history()` method. Store the retrieved data segments in a list.


**Reasoning**:
To retrieve the 2-day stock history for each earnings date, I will iterate through the `earnings_dates_list`, calculate the end date (earnings date + 1 day), fetch the historical data using `Apple.history()`, and append the resulting DataFrame to a list.



```python
historical_data_list = []

for earnings_date in earnings_dates_list:
    # Calculate the end date for a 2-day period (earnings date and the following day)
    end_date_for_period = earnings_date + pd.Timedelta(days=1)

    # Fetch the stock history for the 2-day period
    # Use .strftime('%Y-%m-%d') to ensure proper date format for yfinance
    data_segment = Apple.history(start=earnings_date.strftime('%Y-%m-%d'), end=end_date_for_period.strftime('%Y-%m-%d'))

    # Append the data segment to the list
    if not data_segment.empty:
        historical_data_list.append(data_segment)

print(f"Number of historical data segments retrieved: {len(historical_data_list)}")
print("First historical data segment:\n", historical_data_list[0].head())
```

**Reasoning**:
The previous step successfully extracted individual 2-day historical data segments for each earnings date, handling cases where data might not be available. The next logical step is to combine all these individual DataFrames into a single, comprehensive DataFrame for further analysis.



```python
historical_data_list = []

for earnings_date in earnings_dates_list:
    # Calculate the end date for a 2-day period (earnings date and the following day)
    end_date_for_period = earnings_date + pd.Timedelta(days=1)

    # Fetch the stock history for the 2-day period
    # Use .strftime('%Y-%m-%d') to ensure proper date format for yfinance
    data_segment = Apple.history(start=earnings_date.strftime('%Y-%m-%d'), end=end_date_for_period.strftime('%Y-%m-%d'))

    # Append the data segment to the list if it's not empty
    if not data_segment.empty:
        historical_data_list.append(data_segment)

# Combine all historical data segments into a single DataFrame
combined_historical_df = pd.concat(historical_data_list)

print(f"Number of historical data segments retrieved: {len(historical_data_list)}")
print("First historical data segment:\n", historical_data_list[0].head())
print("\nCombined Historical Data DataFrame:\n", combined_historical_df.head())
print(f"Total rows in combined DataFrame: {len(combined_historical_df)}")
```
